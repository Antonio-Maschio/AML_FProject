# chunks

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import DBSCAN

def load_image(image_path):
    try:
        img = Image.open(image_path)
        img = img.convert('RGB')  # Ensure the image is in RGB format
        return img
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

def extract_red_channel(image):
    try:
        red_channel = np.array(image)[:, :, 0].astype(np.uint8)  # Extract the red channel and ensure it's uint8
        return red_channel
    except Exception as e:
        print(f"Error extracting red channel: {e}")
        return None

def apply_dbscan_chunked(red_channel, chunk_size, eps=0.0001, min_samples=1):
    height, width = red_channel.shape
    labels = np.zeros((height, width), dtype=int) - 1  # Initialize with -1 (unclustered)

    for i in range(0, height, chunk_size):
        for j in range(0, width, chunk_size):
            chunk = red_channel[i:i+chunk_size, j:j+chunk_size]
            red_flat = chunk.flatten().reshape(-1, 1)
            clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(red_flat)
            chunk_labels = clustering.labels_.reshape(chunk.shape)
            labels[i:i+chunk_size, j:j+chunk_size] = chunk_labels

    return labels

def plot_clusters(red_channel, labels):
    try:
        plt.figure(figsize=(12, 6))
        plt.subplot(1,3,1)
        plt.title('Original Image')
        plt.imshow(image)
        plt.subplot(1, 3, 2)
        plt.title('Red Channel')
        plt.imshow(red_channel, cmap='Reds')
        plt.subplot(1, 3, 3)
        plt.title('DBSCAN Clustering')
        plt.imshow(labels, cmap='inferno')
        plt.savefig("firstout.png")

        plt.show()
    except Exception as e:
        print(f"Error plotting clusters: {e}")

if __name__ == "__main__":
    image_path = 'image/control/001_z26.png'
    
    # image_path = 'image/penetramax/011_z57.png'

    chunk_size = 400  # Define the chunk size 400
    image = load_image(image_path)
    
    if image is not None:
        red_channel = extract_red_channel(image)
        
        if red_channel is not None:
            print("Processing image in chunks...")
            labels = apply_dbscan_chunked(red_channel, chunk_size)
            
            if labels is not None:
                plot_clusters(red_channel, labels)
```

```python
import cupy as cp
import matplotlib.pyplot as plt
from PIL import Image
from cuml.cluster import DBSCAN as cumlDBSCAN

def load_image(image_path):
    try:
        img = Image.open(image_path)
        img = img.convert('RGB')  # Ensure the image is in RGB format
        return img
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

def downsample_image(image, scale_factor):
    try:
        width, height = image.size
        new_size = (width // scale_factor, height // scale_factor)
        downsampled_image = image.resize(new_size, Image.ANTIALIAS)
        return downsampled_image
    except Exception as e:
        print(f"Error downsampling image: {e}")
        return None

def extract_red_channel(image):
    try:
        red_channel = cp.array(image)[:, :, 0].astype(cp.float32)  # Extract the red channel and ensure it's uint8
        return red_channel
    except Exception as e:
        print(f"Error extracting red channel: {e}")
        return None

def apply_dbscan(red_channel, eps=0.1, min_samples=2): # 0.1, 5
    try:
        # Flatten the 2D red channel into a 1D array of pixel values
        red_flat = red_channel.flatten().reshape(-1, 1)
        # Apply GPU-accelerated DBSCAN
        clustering = cumlDBSCAN(eps=eps, min_samples=min_samples).fit(red_flat)
        labels = clustering.labels_.reshape(red_channel.shape)
        return labels
    except Exception as e:
        print(f"Error applying DBSCAN: {e}")
        return None

def plot_clusters(red_channel, labels):
    try:
        plt.figure(figsize=(14, 8))
        plt.subplot(1,3,1)
        plt.title('Original')
        plt.imshow(image)
        plt.subplot(1, 3, 2)
        plt.title('Red Channel')
        plt.imshow(cp.asnumpy(red_channel), cmap='Reds')
        plt.subplot(1, 3, 3)
        plt.title('DBSCAN Clustering')
        plt.imshow(cp.asnumpy(labels), cmap='viridis')
        plt.savefig("gpuout.png")
        plt.show()
    except Exception as e:
        print(f"Error plotting clusters: {e}")

if __name__ == "__main__":
    image_path = 'image/control/001_z26.png'  # Update this path to your .png file
    # scale_factor = 2  # Reduce the image size by this factor
    image = load_image(image_path)
    
    if image is not None:
        # image = downsample_image(image, scale_factor)
        red_channel = extract_red_channel(image)
        
        if red_channel is not None:
            print("Processing image with GPU...")
            labels = apply_dbscan(red_channel)
            print(labels)
            
            if labels is not None:
                plot_clusters(red_channel, labels)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('image/control/001_z26.png')

# Define the range for pure red in the image
# Adjust these values if necessary
lower_red = np.array([0, 0, 250])
upper_red = np.array([0, 0, 255])

# Create a mask for pure red
mask = cv2.inRange(image, lower_red, upper_red)

# Find contours (red spots)
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a mask for inpainting
inpaint_mask = np.zeros_like(mask)
cv2.drawContours(inpaint_mask, contours, -1, (255), thickness=cv2.FILLED)

# Inpaint to remove the bright red spots
result = cv2.inpaint(image, inpaint_mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

# Display the original and the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.title('Image after Removing Red Spots')
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
