```python
import numpy as np
import matplotlib.pyplot as plt

from liams_funcs import *
from segmentation import *
from features import *
```

```python
### Demonstration of hist correction

from skimage.exposure import equalize_hist

N_samples = 10_000

peak1 = np.random.normal(loc=50, scale=25, size=N_samples//2).astype(np.uint8)
peak2 = np.random.normal(loc=200, scale=25, size=N_samples).astype(np.uint8)
spectrum1 = np.append(peak1, peak2)
corrected1 = (255 * equalize_hist(spectrum1)).astype(np.uint8)

peak3 = np.random.normal(loc=50, scale=25, size=N_samples).astype(np.uint8)
peak4 = np.random.normal(loc=200, scale=25, size=N_samples//2).astype(np.uint8)
spectrum2 = np.append(peak3, peak4)
corrected2 = (255 * equalize_hist(spectrum2)).astype(np.uint8)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,6), dpi=300, sharex=True)
plt.subplots_adjust(hspace=0)

axes[0,0].set_title('Raw Spectra'); axes[0,1].set_title('Corrected Spectra')

axes[0,0].hist(spectrum1, bins=64, range=(0, 255), color='orangered')
axes[0,1].hist(corrected1, bins=64, range=(0, 255), color='firebrick')

axes[1,0].hist(spectrum2, bins=64, range=(0, 255), color='navy')
axes[1,1].hist(corrected2, bins=64, range=(0, 255), color='darkorchid')

for ax in axes.flatten():
    ax.set_xlim(0, 255)
    ax.set_xlabel('RGB Value'); ax.set_ylabel('Counts')

plt.savefig('../figures/hist_demonstration.png', dpi=300, bbox_inches='tight')
plt.show()
```
