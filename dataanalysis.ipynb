```python
def plot_all_histograms(data_path, n_bins=10, figsize=(12, 8)):
  """
  Plots histograms for all numerical columns in a pandas DataFrame.

  Args:
    data_path (str): Path to the CSV file containing the data.
    n_bins (int, optional): Number of bins to use for the histograms. Defaults to 10.
    figsize (tuple, optional): Size of the figure for the plots. Defaults to (12, 8).
  """
  df = pd.read_csv(data_path)
  numerical_cols = df.select_dtypes(include=[np.number])  # Select numerical columns

  # Create a subplot grid to accommodate multiple histograms
  fig, axes = plt.subplots(nrows=len(numerical_cols) // 3 + 1, ncols=3, figsize=figsize)

  # Iterate through numerical columns and plot histograms on subplots
  col_index = 0
  row_index = 0
  for col in numerical_cols:
    axes[row_index, col_index].hist(df[col], bins=n_bins)
    axes[row_index, col_index].set_title(col)
    axes[row_index, col_index].set_xlabel(col)
    axes[row_index, col_index].set_ylabel("Frequency")

    col_index += 1
    if col_index == 3:
      col_index = 0
      row_index += 1

  # Adjust layout to remove empty subplots
  plt.tight_layout()
  plt.show()

# Example usage
plot_all_histograms("features.csv", n_bins=15, figsize=(15, 10))
```

```python
import os
def plot_all(image_list,until):
    images = image_list[:until]
    binary_nucleus_images=[]
    for i,im in enumerate(images):
        binary_nucleus_images.append(im)
            # If we have accumulated 6 images, display them
        if len(binary_nucleus_images) == 6:

            fig, ax = plt.subplots(1, 6, figsize=(18, 3)) 
            
            for o in range(6):
                ax[o].imshow(binary_nucleus_images[o], cmap='gray')
                ax[o].axis('off')  # Hide the axes for better visualization
            fig.text(0.0, 0.5, f'Nucleus Numbers from {i - 5} to {i}', va='center', rotation='vertical', fontsize=14)
            plt.tight_layout()
            plt.show()
            
            # Clear the list for the next set of images
            binary_nucleus_images.clear()
def load_images_from_folder(folder_path):
    from skimage.io import imread
    import os
    """
    Load all images from the specified folder and return them as a list of images read by OpenCV.
    """
    images = []
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        img = imread(image_path)
        if img is not None:
            images.append(img)
        else:
            print(f"Error loading image {filename}")
    return images

# Read masks
def load_npy_files(base_dir):
    subfolders = ['control', 'penetramax']
    
    outputs = []
    
    for subfolder in subfolders:
        subfolder_path = os.path.join(base_dir, subfolder)
        
        if os.path.exists(subfolder_path):
            for filename in os.listdir(subfolder_path):
                if filename.endswith('.npy'):
                    file_path = os.path.join(subfolder_path, filename)
                    data = np.load(file_path)
                    outputs.append(data)
        else:
            print(f"Subfolder {subfolder_path} does not exist.")
    return outputs

def get_masks(base_path):
    segments = load_npy_files(base_path)
    segments_nuclei = []
    for segment in segments:
        segments_nuclei.append(segment[:, :, 1])
    return segments_nuclei[:10],segments_nuclei[10:]

# Find median image
def get_median_image(ctrl_masks,pntr_masks):
    from skimage import measure
    ctrl_temp = []
    pntr_temp = []
    for k,i in enumerate(ctrl_masks):
        regions = measure.regionprops(i)
        for j,reg in enumerate(regions):
            bi_mask = reg.image
            ctrl_temp.append(bi_mask.shape)
    for k,i in enumerate(pntr_masks):
        regions = measure.regionprops(i)
        for j,reg in enumerate(regions):
            bi_mask = reg.image
            pntr_temp.append(bi_mask.shape)

    ctrl_temp = np.array(ctrl_temp)
    pntr_temp = np.array(pntr_temp)
    
    max_ctrl = np.median(ctrl_temp, axis=0)
    max_pntr = np.median(pntr_temp, axis=0)
    
    return int(np.median([max_ctrl, max_pntr]))

#Get each cell in an image list
def get_cells_re(image_list,mask_list):
    from skimage import measure
    temp = []
    for k,i in enumerate(mask_list):
        regions = measure.regionprops(i)
        for j,reg in enumerate(regions):
            min_row, min_col, max_row, max_col = reg.bbox

            region_of_interest = image_list[k][min_row:max_row,min_col:max_col]
            
            temp.append(region_of_interest)
    return temp
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import umap
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
# from clusteval import clusteval

# Load the dataset
df = pd.read_csv('features.csv')

# Define the variables to be used for clustering
variables = ['getAreaOfHoles', 'getCellArea', 'getCellAreaBbox', 'getCellAreaConvex', 'getCellAxisMajor',
             'getCellAxisMinor', 'getCellDiam', 'getCellEcc', 'getCellFeretDiam', 'getCellMaxR',
             'getCellMinR', 'getCellOri', 'getCellPerim', 'getCellRoundness', 'getCellSolidity',
             'getEdgeFluxR', 'getEnergyB', 'getEnergyR', 'getEntropyB',
             'getEntropyR', 'getEuler', 'getH10B', 'getH10R', 'getH11B', 'getH11R',
             'getH12B', 'getH12R', 'getH13B', 'getH13R', 'getH1B', 'getH2B',
             'getH2R', 'getH3B', 'getH3R', 'getH4B', 'getH4R', 'getH5B', 'getH5R',
             'getH6B', 'getH6R', 'getH7B', 'getH7R', 'getH8B', 'getH8R', 'getH9B',
             'getH9R', 'getNHoles', 'getNObjects', 'getNucleusArea', 'getNucleusAreaBbox',
             'getNucleusAreaConvex', 'getNucleusAxisMajor', 'getNucleusAxisMinor', 'getNucleusDiam', 'getNucleusEcc',
             'getNucleusFeretDiam', 'getNucleusFraction', 'getNucleusMaxR', 'getNucleusMinR', 'getNucleusOri',
             'getNucleusPerim', 'getNucleusRoundness', 'getNucleusSolidity', 'getSegmentOrder']

# Extract the variables for clustering
data = df[variables].values

# Normalize the data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Use NearestNeighbors to calculate the k-distance graph
k = 100  # Typically, we choose min_samples = k
nearest_neighbors = NearestNeighbors(n_neighbors=k)
neighbors = nearest_neighbors.fit(data_normalized)
distances, indices = neighbors.kneighbors(data_normalized)

# Sort the distances (kth nearest neighbor distance)
distances = np.sort(distances[:, k-1])

# Plot the k-distance graph
plt.figure(figsize=(10, 6))
plt.plot(distances[2000:])
plt.title('k-distance graph')
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{k}th Nearest Neighbor Distance')
plt.show()
```
